#### `ceiling_height`

Минимальная высота потолка в датафрейме - 1 метр, максимальная - 100 метров. Также пропущена значительная часть данных о высоте потолков. Изучим распределение высот потолков:

Количество объектов недвижимости с высотой потолка более 5 метров невелико. На основании площадей и цен можем предположить, что какие-то из этих данных - корректны, проверяем:

```python
df_overheight = df.query('ceiling_height >= 5')
df_overheight.sort_values(by='last_price')
```

Изучив полученные данные, можем предположить следующее:
- потолки до 6.0 метров возможны либо в больших и дорогих домах, либо в нежелых помещениях;
- потолки от 22.0 до 32.0 метров разделим на 10. Возможно, при вводе произошла опечатка с расположением точки. Полученные значения будут распологаться в пределах нормальных значений, то есть "усов" графика размаха;
- остальные значения удаляем как некорректные.

```python
# Убеждаемся, что в графике размаха действительно основные значения лежат примерно в пределах от 2.2 до 3.2 метров
df.query('ceiling_height <= 5').boxplot(column='ceiling_height')
```

```python
def ceiling_height_normalizer(ceiling_height):
    if ceiling_height < 2.2: # разносим три диапозона с return 0 в разные условия, если позднее придётся какой-либо изменять
        return np.nan
    elif 6 < ceiling_height < 22:
        return np.nan
    elif 22 <= ceiling_height <= 32:
        return ceiling_height / 10
    elif ceiling_height > 32:
        return np.nan
    else:
        return ceiling_height

df['ceiling_height'] = df['ceiling_height'].apply(ceiling_height_normalizer)

df['ceiling_height'].value_counts()
```

```python
# Проверяет полученные данные
df.boxplot(column='ceiling_height');
```

Так как мы не можем корректно отфильтровать данные от снятых без продажи объявлений, присвоим всем значениям 30, 45, 60 и 90 в столбце `days_exposition` значение NaN и сохраним данные в отдельный датафрейм.
